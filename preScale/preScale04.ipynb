```python
df1 = df1[['날짜','시군구','읍면동','방문자수']]
df1.columns = ['날짜','시군구','읍면동','내국인방문자수']
```

```python
a = df1.groupby(['날짜','시군구','읍면동']).sum()
a
```

```python
df2 = df2[['날짜','시군구','읍면동','방문자수']]
df2.columns = ['날짜','시군구','읍면동','단기체류외국인방문자수']
```

```python
df2
```

```python
b = df2.groupby(['날짜','시군구','읍면동']).sum()
b
```

```python
target = pd.merge(left = a , right = b, how = "left", on = ["날짜","시군구","읍면동"])
target
```

```python
df3 = df3[['날짜','시군구','읍면동','방문자수']]
df3.columns = ['날짜','시군구','읍면동','장기체류외국인방문자수']
```

```python
c = df3.groupby(['날짜','시군구','읍면동']).sum()
c
```

```python
target = pd.merge(left=target, right=c, how="left", on = ["날짜","시군구","읍면동"])
target
```
